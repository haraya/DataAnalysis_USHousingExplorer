```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
data = pd.read_json('HPI_master.json')

sortYears = data.sort_values(by='yr')

soldHouses_1975 =len(sortYears[sortYears['yr']==1975])
soldHouses_1985 =len(sortYears[sortYears['yr']==1985])
soldHouses_2000 = len(sortYears[sortYears['yr']==2000])
soldHouses_2023 = len(sortYears[sortYears['yr']==2023])

yearsHouses = ["1975","2000","2023"]
valuesHouses= [soldHouses_1975, soldHouses_2000, soldHouses_2023]

#124,814 data
```

```python
plt.bar(yearsHouses, valuesHouses)

for i, value in enumerate(valuesHouses):
    plt.text(i,value+0.5, str(value), ha='center', va='bottom')

plt.xlabel("Years")
plt.ylabel("Number of houses sold ")
plt.title("House Sales Distribution: 25-Year Periods")
plt.show()
```

```python
quarterly = len(data[data['frequency']=='quarterly'])
monthly = len(data[data['frequency']=='monthly'])

frequencyValues = [monthly, quarterly]
frequencyText = ['Montly', 'Quarterly'] 

plt.pie(frequencyValues, labels=frequencyText, autopct="%1.1f%%")
plt.title("Monthly vs. Quarterly: Dominating Sales Frequency?")
plt.show()

print(f"""\tHouses Sold for quarterly = 98.6% ({quarterly}),\tHouses Sold for monthly = 3.2% ({monthly}) """)
```

```python
yearsList = list(set(sortYears['yr']))

dictSoldHouses = {}
for year in yearsList:
    soldHouses =len(sortYears[sortYears['yr']==year])
    dictSoldHouses[year] = soldHouses

yearsHouses_before_2000 = []
valuesHouses_before_2000= []

yearsHouses_after_2000 = []
valuesHouses_after_2000= []

for key, value in dictSoldHouses.items():
    if key <2000:
        yearsHouses_before_2000.append(str(key))
        valuesHouses_before_2000.append(value)
    if key >=2000:
        yearsHouses_after_2000.append(str(key))
        valuesHouses_after_2000.append(value)
        
```

```python
#Bar chart of Houses sold before the year 2000
fig, axes = plt.subplots(1,2, figsize=(13,10))

bars1 = axes[0].barh(yearsHouses_before_2000, valuesHouses_before_2000, color='#41B06E')

for i, value in enumerate(valuesHouses_before_2000):
    axes[0].text(value + 0.5, i - 0.1, f"{value}", ha='left', va='center', fontsize=12)
    
axes[0].set_ylabel("Years")
axes[0].set_xlabel("Number of houses sold ")
axes[0].set_title("Houses sold before the year 2000.")





#-----------------------Horizontal Bar chart After 2000
#Bar chart of Houses sold after the year 2000
years = yearsHouses_after_2000
values = valuesHouses_after_2000

bars2 =axes[1].barh(years, values, color='skyblue')

for i, value in enumerate(values):
    axes[1].text(value + 0.5, i - 0.1, f"{value}", ha='left', va='center', fontsize=12)

axes[1].set_ylabel("Years")
axes[1].set_xlabel("Number of houses sold ")
axes[1].set_title("Houses sold after the year 2000.")

plt.tight_layout()
plt.show()






```

```python
# Bar chart about sold houses by period
periodList = list(set(sortYears['period']))

dictPeriodSoldHouses = {}

for period in periodList:
    periodHouses =len(sortYears[sortYears['period']==period])
    dictPeriodSoldHouses[period] = periodHouses


listPeriodText = []
listPeriodValues = []
for keyPeriod, valuePeriod in dictPeriodSoldHouses.items():
    listPeriodText.append(str(keyPeriod))
    listPeriodValues.append(valuePeriod)


# Create the bar
plt.bar(listPeriodText,listPeriodValues)

# Add annotations above each bar with period and value
for i, (x, y) in enumerate(zip(listPeriodText, listPeriodValues)):
    plt.annotate(f"{y}", (x, y + 0.1), ha='center', va='bottom', fontsize=8)

# Customize the plot (optional)
plt.ylabel('Number of Houses Sold')
plt.xlabel('Period')
plt.title('Boom or Bust? Unveiling the Distribution of Houses Sold by Period')
plt.show()
```
